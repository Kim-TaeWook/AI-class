```python
# 1️⃣ 라이브러리 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler  # 데이터 표준화

# 2️⃣ 데이터 로드
file_path = r"C:\Users\hak87\Downloads\3학년 1학기\3.화_인공지능개론\week2\mobile.csv"  # 파일 경로
df = pd.read_csv(file_path)

# 데이터프레임 확인
print(df.head())  # 첫 5개 행 출력
print(df.columns)  # 컬럼 확인

# 3️⃣ 데이터 전처리 (입력 데이터 X, 정답 y 설정)
X = df.drop(columns=["price_range"])  # 'price_range' 제외한 나머지 (특성 데이터)
y = df["price_range"]  # price_range가 정답(label)

# 4️⃣ 데이터 표준화 (로지스틱 회귀 모델 성능 향상을 위해)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # 모든 데이터 스케일링 적용

# 5️⃣ 데이터 분할 (학습용 80%, 테스트용 20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 6️⃣ 결정 트리 (Decision Tree)
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)

# 7️⃣ 랜덤 포레스트 (Random Forest)
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)

# 8️⃣ 서포트 벡터 머신 (SVM)
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)

# 9️⃣ 로지스틱 회귀 (Logistic Regression)
lr_model = LogisticRegression(max_iter=1000)  # max_iter 증가
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)

# 🔟 결과 출력
print(f"결정 트리 정확도: {dt_acc:.4f}")
print(f"랜덤 포레스트 정확도: {rf_acc:.4f}")
print(f"서포트 벡터 머신 정확도: {svm_acc:.4f}")
print(f"로지스틱 회귀 정확도: {lr_acc:.4f}")
```
