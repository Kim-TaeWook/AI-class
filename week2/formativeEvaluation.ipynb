```python
# 1ï¸âƒ£ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler  # ë°ì´í„° í‘œì¤€í™”

# 2ï¸âƒ£ ë°ì´í„° ë¡œë“œ
file_path = r"C:\Users\hak87\Downloads\3í•™ë…„ 1í•™ê¸°\3.í™”_ì¸ê³µì§€ëŠ¥ê°œë¡ \week2\mobile.csv"  # íŒŒì¼ ê²½ë¡œ
df = pd.read_csv(file_path)

# ë°ì´í„°í”„ë ˆì„ í™•ì¸
print(df.head())  # ì²« 5ê°œ í–‰ ì¶œë ¥
print(df.columns)  # ì»¬ëŸ¼ í™•ì¸

# 3ï¸âƒ£ ë°ì´í„° ì „ì²˜ë¦¬ (ì…ë ¥ ë°ì´í„° X, ì •ë‹µ y ì„¤ì •)
X = df.drop(columns=["price_range"])  # 'price_range' ì œì™¸í•œ ë‚˜ë¨¸ì§€ (íŠ¹ì„± ë°ì´í„°)
y = df["price_range"]  # price_rangeê°€ ì •ë‹µ(label)

# 4ï¸âƒ£ ë°ì´í„° í‘œì¤€í™” (ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ ì„±ëŠ¥ í–¥ìƒì„ ìœ„í•´)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # ëª¨ë“  ë°ì´í„° ìŠ¤ì¼€ì¼ë§ ì ìš©

# 5ï¸âƒ£ ë°ì´í„° ë¶„í•  (í•™ìŠµìš© 80%, í…ŒìŠ¤íŠ¸ìš© 20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 6ï¸âƒ£ ê²°ì • íŠ¸ë¦¬ (Decision Tree)
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)

# 7ï¸âƒ£ ëœë¤ í¬ë ˆìŠ¤íŠ¸ (Random Forest)
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)

# 8ï¸âƒ£ ì„œí¬íŠ¸ ë²¡í„° ë¨¸ì‹  (SVM)
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)

# 9ï¸âƒ£ ë¡œì§€ìŠ¤í‹± íšŒê·€ (Logistic Regression)
lr_model = LogisticRegression(max_iter=1000)  # max_iter ì¦ê°€
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)

# ğŸ”Ÿ ê²°ê³¼ ì¶œë ¥
print(f"ê²°ì • íŠ¸ë¦¬ ì •í™•ë„: {dt_acc:.4f}")
print(f"ëœë¤ í¬ë ˆìŠ¤íŠ¸ ì •í™•ë„: {rf_acc:.4f}")
print(f"ì„œí¬íŠ¸ ë²¡í„° ë¨¸ì‹  ì •í™•ë„: {svm_acc:.4f}")
print(f"ë¡œì§€ìŠ¤í‹± íšŒê·€ ì •í™•ë„: {lr_acc:.4f}")
```
